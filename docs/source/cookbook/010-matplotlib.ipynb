# MatplotLib

Note: Matplotlib is not natively installed within Molli, but it can be added through conda using the line:
`pip install matplotlib` or `conda install matplotlib`

```python
# Import necessary packages
import molli as ml 
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Loads Pentane molecule from mol2 file
fpath = ml.files.pentane_confs_mol2
pentane_mol = ml.Molecule.load_mol2(fpath, name="pentane")

pentane_mol = pentane_mol

# Extracts the symbols of the atoms in the molecule

symbols = []
for i in range(pentane_mol.n_atoms):
    symbols.append(pentane_mol.atoms[i].element.symbol)
    print(pentane_mol.atoms[i].element.symbol)

```

```python
# Create a figure and a 3D axes
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Extract x, y, z coordinates
x_coords = pentane_mol.coords[:, 0]
y_coords = pentane_mol.coords[:, 1]
z_coords = pentane_mol.coords[:, 2]

# Extract atomic symbols
color_key = {'C': 'orange', 'H': 'blue'}


# Set labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Molecule Visualization')

# Adjust the viewing angle
ax.view_init(elev=30, azim=45)

# Plots the scatter plot
for i in range(pentane_mol.n_atoms):
    ax.scatter(x_coords[i], y_coords[i], z_coords[i],
               color=color_key[symbols[i]], s=50, label=symbols[i])

# Shows the plot
plt.show()

```
